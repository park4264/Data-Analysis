## 3-2. í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§ II: ë°ì´í„° ë‹¤ìš´ìºìŠ¤íŒ…

- **ë‹¤ìš´ìºìŠ¤íŒ…(downcasting)**: ë” ì‘ì€ ë°ì´í„°íƒ€ì…ìœ¼ë¡œ ë³€í™˜í•˜ëŠ” ì‘ì—…
- ë°ì´í„° ë‹¤ìš´ìºìŠ¤íŒ…ì„ ì™œ í•˜ëŠ”ì§€ì™€ ê·¸ ë°©ë²•ì„ ë°°ì›Œë³´ì


- íŒë‹¤ìŠ¤ë¡œ ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¤ë©´ ê¸°ë³¸ì ìœ¼ë¡œ ì •ìˆ˜í˜•ì€ `int64`, ì‹¤ìˆ˜í˜•ì€ `float64` íƒ€ì…ì„ í• ë‹¹
- ê°ê° ì •ìˆ˜í˜•, ì‹¤ìˆ˜í˜•ì—ì„œ ê°€ì¥ í° íƒ€ì…
- í¬ì§€ ì•Šì€ ìˆ«ìê°€ ì €ì¥ëœ í”¼ì²˜ë¼ë©´ `int8`, `int16`, `int32`, `float16`, `float32` ë“± ë³´ë‹¤ ì‘ì€ íƒ€ì…ìœ¼ë¡œ í• ë‹¹í•  ìˆ˜ ìˆë‹¤.
    - ë©”ëª¨ë¦¬ ë‚­ë¹„ë¥¼ ë§‰ê³ , í›ˆë ¨ ì†ë„ë¥¼ ë†’í ìˆ˜ ìˆìŒ
    



```python
# í•´ë‹¹ í”¼ì²˜ í¬ê¸°ì— ë§ê²Œ ì ì ˆí•œ íƒ€ì…ìœ¼ë¡œ ë°”ê¿”ì£¼ëŠ” í•¨ìˆ˜
def downcast(df, verbose=True):
    start_mem = df.memory_usage().sum() / 1024**2
    for col in df.columns:
        dtype_name = df[col].dtype.name
        if dtype_name == 'object':
            pass
        elif dtype_name == 'bool':
            df[col] = df[col].astype('int8')
        elif dtype_name.startswith('int') or (df[col].round() == df[col]).all():
            df[col] = pd.to_numeric(df[col], downcast='integer')
        else:
            df[col] = pd.to_numeric(df[col], downcast='float')
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose:
        print('{:.1f}% ì••ì¶•ë¨'.format(100 * (start_mem - end_mem) / start_mem))
    
    return df
```

```python
all_df = [sales_train, shops, items, item_categories, test]
for df in all_df:
    df = downcast(df)
```

## 3-3. í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§ III : ë°ì´í„° ì¡°í•© ìƒì„±

- ìš°ë¦¬ëŠ” `ì›”ID`, `ìƒì ID`, `ìƒí’ˆID` í”¼ì²˜ ì¡°í•©ì´ 
- ë°ì´í„° ì¡°í•©ì€ `itertools`ê°€ ì œê³µí•˜ëŠ” `product()` í•¨ìˆ˜ë¡œ ì‰½ê²Œ ë§Œë“¤ì–´ë‚¼ ìˆ˜ ìˆë‹¤.

```python
from itertools import product

train = []
# ì›”ID, ìƒì ID, ìƒí’ˆID ì¡°í•© ìƒì„± 
for i in sales_train['ì›”ID'].unique():
    all_shop = sales_train.loc[sales_train['ì›”ID']==i, 'ìƒì ID'].unique()
    all_item = sales_train.loc[sales_train['ì›”ID']==i, 'ìƒí’ˆID'].unique()
    train.append(np.array(list(product([i], all_shop, all_item))))

idx_features = ['ì›”ID', 'ìƒì ID', 'ìƒí’ˆID'] # ê¸°ì¤€ í”¼ì²˜
# ë¦¬ìŠ¤íŠ¸ íƒ€ì…ì¸ trainì„ DataFrame íƒ€ì…ìœ¼ë¡œ ë³€í™˜ 
train = pd.DataFrame(np.vstack(train), columns=idx_features)

train
```

- ì›” IDì˜ ê³ ìœ³ê°’ (0~33) ë³„ë¡œ ëª¨ë“  ìƒì ID ê³ ìœ³ê°’, ìƒí’ˆID ê³ ìœ³ê°’ì„ êµ¬í•´ ì¡°í•©ì„ ìƒì„±
- ì´ë ‡ê²Œ ë§Œë“  `train`ì„ ì•ìœ¼ë¡œ í›ˆë ¨ ë°ì´í„°ì˜ ë¼ˆëŒ€ë¡œ ì‚¬ìš©í• ê²ƒì´ë‹¤.
- ë¼ˆëŒ€ê°€ ë˜ëŠ” `train`ì— íƒ€ê¹ƒê°’ì„ ë³‘í•©í•˜ê³ , ë‚˜ë¨¸ì§€ `shops.csv`, `items.csv`, `item_categories.csv`ë„ ë³‘í•©í•  ê²ƒì´ë‹¤.

## 3-4. í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§ IV : íƒ€ê¹ƒê°’(ì›”ê°„ íŒë§¤ëŸ‰) ì¶”ê°€

- `sales_train`ì—ëŠ” <u>ì¼ë³„ íŒë§¤ëŸ‰</u>ì„ ë‚˜íƒ€ë‚´ëŠ” 'íŒë§¤ëŸ‰' í”¼ì²˜ê°€ ìˆë‹¤
- í•˜ì§€ë§Œ ìš°ë¦¬ê°€ ì›í•˜ëŠ” íƒ€ê¹ƒê°’ì€ ê° ìƒì ì˜ ìƒí’ˆë³„ <u>ì›”ê°„ íŒë§¤ëŸ‰</u>ì…ë‹ˆë‹¤.
- ì´ ê°’ì„ êµ¬í•˜ë ¤ë©´ `ì›”ID`, `ìƒì ID`, `ìƒí’ˆID`ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ê·¸ë£¹í™”í•´ íŒë§¤ëŸ‰ì„ ë”í•´ì•¼ í•©ë‹ˆë‹¤.
    - `groupby()` í•¨ìˆ˜ë¥¼ ì´ìš©

```python
# idx_featuresë¥¼ ê¸°ì¤€ìœ¼ë¡œ ê·¸ë£¹í™”í•´ íŒë§¤ëŸ‰ í•© êµ¬í•˜ê¸° 
group = sales_train.groupby(idx_features).agg({'íŒë§¤ëŸ‰': 'sum'})
# ì¸ë±ìŠ¤ ì¬ì„¤ì •
group = group.reset_index()
# í”¼ì²˜ëª…ì„ 'íŒë§¤ëŸ‰'ì—ì„œ 'ì›”ê°„ íŒë§¤ëŸ‰'ìœ¼ë¡œ ë³€ê²½
group = group.rename(columns={'íŒë§¤ëŸ‰': 'ì›”ê°„ íŒë§¤ëŸ‰'})

group
```

- `train`ê³¼ `group`ì„ ë³‘í•©í•´ë³´ì
- `train`ì€ ì›”ID, ìƒì ID, ìƒí’ˆID, ì¡°í•©ì´ë¯€ë¡œ `group`ì„ ë³‘í•©í•˜ë©´
    - â¡ï¸ ì›”ID, ìƒì ID, ìƒí’ˆID, ì›”ê°„ íŒë§¤ëŸ‰ ì¡°í•©ì„ êµ¬í•  ìˆ˜ ìˆë‹¤.

```python
# trainê³¼ group ë³‘í•©í•˜ê¸°
train = train.merge(group, on=idx_features, how='left')

train
```

- ê²°ì¸¡ê°’ì€ 0ìœ¼ë¡œ ì¶”í›„ì— ëŒ€ì²´í•  ì˜ˆì •

### ğŸ—‘ï¸ ê°€ë¹„ì§€ ì»¬ë ‰ì…˜(Garbage collection)

- `group` ë°ì´í„°ëŠ” ë” ì´ìƒ í•„ìš” ì—†ìœ¼ë‹ˆ ë©”ëª¨ë¦¬ ì ˆì•½ ì°¨ì›ì—ì„œ ê°€ë¹„ì§€ ì»¬ë ‰ì…˜ì„ í•´ì£¼ê² ë‹¤.

```python
import gc # ê°€ë¹„ì§€ ì»¬ë ‰í„° ë¶ˆëŸ¬ì˜¤ê¸°

del group # ë”ëŠ” ì‚¬ìš©í•˜ì§€ ì•ŠëŠ” ë³€ìˆ˜ ì§€ì •
gc.collect(); # ê°€ë¹„ì§€ ì»¬ë ‰ì…˜ ìˆ˜í–‰
```

## 3-5. í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§ V : í…ŒìŠ¤íŠ¸ ë°ì´í„° ì´ì–´ë¶™ì´ê¸°

- ì´ì œ `train` ë°ì´í„°ì— í…ŒìŠ¤íŠ¸ ë°ì´í„°(`test`)ë¥¼ ì´ì–´ë¶™ì¼ ê²ƒì´ë‹¤.
- í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¥¼ ì´ì–´ë¶™ì´ëŠ” ì´ìœ ëŠ” ë’¤ì´ì–´ `shops.csv`, `items.csv`, `item_categories` ë°ì´í„°ë¥¼ ë³‘í•©í•  ë•Œ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë„ í•œë²ˆì— í•˜ê¸° ìœ„í•´ì„œì´ë‹¤.

```python
test['ì›”ID'] = 34
```

```python
# trainê³¼ test ì´ì–´ë¶™ì´ê¸°
all_data = pd.concat([train, test.drop('ID', axis=1)],
                     ignore_index=True, # ê¸°ì¡´ ì¸ë±ìŠ¤ ë¬´ì‹œ(0ë¶€í„° ìƒˆë¡œ ì‹œì‘)
                     keys=idx_features) # ì´ì–´ë¶™ì´ëŠ” ê¸°ì¤€ì´ ë˜ëŠ” í”¼ì²˜
```

```python
# ê²°ì¸¡ê°’ì„ 0ìœ¼ë¡œ ëŒ€ì²´
all_data = all_data.fillna(0)

all_data
```

## 3-6. í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§ VI : ë‚˜ë¨¸ì§€ ë°ì´í„° ë³‘í•©(ìµœì¢… ë°ì´í„° ìƒì„±)

```python
# ë‚˜ë¨¸ì§€ ë°ì´í„° ë³‘í•©
all_data = all_data.merge(shops, on='ìƒì ID', how='left')
all_data = all_data.merge(items, on='ìƒí’ˆID', how='left')
all_data = all_data.merge(item_categories, on='ìƒí’ˆë¶„ë¥˜ID', how='left')

# ë°ì´í„° ë‹¤ìš´ìºìŠ¤íŒ…
all_data = downcast(all_data)

# ê°€ë¹„ì§€ ì»¬ë ‰ì…˜
del shops, items, item_categories
gc.collect();
```

```python
all_data.head()
```

```python
all_data = all_data.drop(['ìƒì ëª…', 'ìƒí’ˆëª…', 'ìƒí’ˆë¶„ë¥˜ëª…'], axis=1)
```

## 3-7. í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§ VII : ë§ˆë¬´ë¦¬

- ì•ì„œ ëª¨ë“  ë°ì´í„°ë¥¼ ë³‘í•©í•´ `all_data`ë¥¼ ë§Œë“¤ì—ˆë‹¤.
- ì´ì œ `all_data`ë¥¼ í™œìš©í•´ í›ˆë ¨, ê²€ì¦, í…ŒìŠ¤íŠ¸ìš© ë°ì´í„°ë¥¼ ë§Œë“¤ ê²ƒì´ë‹¤.
>    - **í›ˆë ¨ ë°ì´í„°**: 2013.01 ~ 2015.09 (ì›”ID = 32) ê¹Œì§€ì˜ íŒë§¤ ë‚´ì—­
>    - **ê²€ì¦ ë°ì´í„°**: 2015.10 (ì›”ID = 33) íŒë§¤ ë‚´ì—­
>    - **í…ŒìŠ¤íŠ¸ ë°ì´í„°**: 2015.11 (ì›”ID = 34) íŒë§¤ ë‚´ì—­

```python
# í›ˆë ¨ ë°ì´í„° (í”¼ì²˜)
X_train = all_data[all_data['ì›”ID'] < 33]
X_train = X_train.drop(['ì›”ê°„ íŒë§¤ëŸ‰'], axis=1)
# ê²€ì¦ ë°ì´í„° (í”¼ì²˜)
X_valid = all_data[all_data['ì›”ID'] == 33]
X_valid = X_valid.drop(['ì›”ê°„ íŒë§¤ëŸ‰'], axis=1)
# í…ŒìŠ¤íŠ¸ ë°ì´í„° (í”¼ì²˜)
X_test = all_data[all_data['ì›”ID'] == 34]
X_test = X_test.drop(['ì›”ê°„ íŒë§¤ëŸ‰'], axis=1)

# í›ˆë ¨ ë°ì´í„° (íƒ€ê¹ƒê°’)
y_train = all_data[all_data['ì›”ID'] < 33]['ì›”ê°„ íŒë§¤ëŸ‰']
y_train = y_train.clip(0, 20) # íƒ€ê¹ƒê°’ì„ 0 ~ 20ë¡œ ì œí•œ
# ê²€ì¦ ë°ì´í„° (íƒ€ê¹ƒê°’)
y_valid = all_data[all_data['ì›”ID'] == 33]['ì›”ê°„ íŒë§¤ëŸ‰']
y_valid = y_valid.clip(0, 20)
```

- ì¶”ê°€ë¡œ `clip()` ë„˜íŒŒì´ í•¨ìˆ˜ë¥¼ í™œìš©í•´ íƒ€ê¹ƒê°’ì¸ 'ê° ìƒì  ìƒí’ˆë³„ ì›”ê°„ íŒë§¤ëŸ‰'ì„ 0~20 ì‚¬ì´ë¡œ ì œí•œ

```python
del all_data
gc.collect();
```

## 3-8. ëª¨ë¸ í›ˆë ¨ ë° ì„±ëŠ¥ ê²€ì¦

- ë² ì´ìŠ¤ë¼ì¸ ëª¨ë¸ë¡œëŠ” "LigthGBM"ì„ ì‚¬ìš©
- ê¸°ë³¸ íŒŒë¼ë¯¸í„°ë§Œ ì„¤ì •í•˜ê³  "LightGBM"ìš© ë°ì´í„°ì…‹ì„ ë§Œë“¤ì–´ì„œ í›ˆë ¨
- `train()` ë©”ì„œë“œì˜ `categorical_feature` íŒŒë¼ë¯¸í„°ì—ëŠ” ë²”ì£¼í˜• ë°ì´í„°ë¥¼ ì „ë‹¬
    - ë²”ì£¼í˜• ë°ì´í„°ë¡œëŠ” ìƒì ID, ìƒí’ˆID, ìƒí’ˆë¶„ë¥˜IDê°€ ìˆë‹¤.
    - ì´ì¤‘ ìƒí’ˆIDë¥¼ ëº€ ìƒì IDì™€ ìƒí’ˆë¶„ë¥˜IDë§Œ ì¸ìˆ˜ë¡œ ì „ë‹¬í•  ê²ƒì´ë‹¤.
    - LightGBM ë¬¸ì„œì— ë”°ë¥´ë©´ ê³ ìœ³ê°’ ê°œìˆ˜ê°€ ë„ˆë¬´ ë§ì€ ë²”ì£¼í˜• ë°ì´í„°ëŠ” ìˆ˜ì¹˜í˜• ë°ì´í„°ë¡œ ì·¨ê¸‰í•´ì•¼ ì„±ëŠ¥ì´ ë” ì˜ ë‚˜ì˜¨ë‹¤ê³  í•œë‹¤.

```python
import lightgbm as lgb

# LightGBM í•˜ì´í¼íŒŒë¼ë¯¸í„°
params = {'metric': 'rmse', # í‰ê°€ì§€í‘œ = rmse
          'num_leaves': 255,
          'learning_rate': 0.01,
          'force_col_wise': True,
          'random_state': 10}

# ë²”ì£¼í˜• í”¼ì²˜ ì„¤ì •
cat_features = ['ìƒì ID', 'ìƒí’ˆë¶„ë¥˜ID']

# LightGBM í›ˆë ¨ ë° ê²€ì¦ ë°ì´í„°ì…‹
dtrain = lgb.Dataset(X_train, y_train)
dvalid = lgb.Dataset(X_valid, y_valid)

# LightGBM ëª¨ë¸ í›ˆë ¨
lgb_model = lgb.train(params=params,
                      train_set=dtrain,
                      num_boost_round=500,
                      valid_sets=(dtrain, dvalid),
                      categorical_feature=cat_features,
                      verbose_eval=50)  
```

## 3-9. ì˜ˆì¸¡ ë° ê²°ê³¼ ì œì¶œ

```python
# ì˜ˆì¸¡
preds = lgb_model.predict(X_test).clip(0, 20)
# ì œì¶œ íŒŒì¼ ìƒì„±
submission['item_cnt_month'] = preds
submission.to_csv('submission.csv', index=False)
```

```python
del X_train, y_train, X_valid, y_valid, X_test, lgb_model, dtrain, dvalid
gc.collect();
```
