## 4-1. 데이터 종류

- 수치형 데이터 (사칙 연산이 가능한 데이터)
    - 연속형 데이터
        - 키, 몸무게, 수입 등
    - 이산형 데이터
        - 과일 개수, 책의 페이지 수 등
        
- 범주형 데이터 (범주로 나누어지는 데이터)
    - 순서형 데이터
        - 학점, 순위(랭킹)
    - 명목형 데이터
        - 성별, 음식 종류, 우편 번호
        
        
## 4-2. 탐색적 데이터 분석과 그래프

- 다양한 그래프를 그린다 $\rightarrow$ 데이터를 한눈에 파악하는데 도움을 줌
    - 데이터 어떻게 구성되있는지
    - 어떤 피처가 중요한지
    - 어떤 피처를 제거할지
    - 어떻게 새로운 피처를 만들지
    - 모델링에 필요한 다양한 정보를 준다.
    
- 코드 예제는 `matplotlib`이 아닌 `seaborn`을 활용함
- `seaborn`: 더 간결, 기본 설정에서의 그래프가 좀 더 정결함



## 4-3. 수치형 데이터 시각화
- 수치형 데이터는 일정한 범위 내에서 어떻게 분포(distribution)되어 있는지가 중요하다.

```python
import seaborn as sns
tips = sns.load_dataset('tips') # 팁 데이터 불러오기
```

```python
tips.head()
```

다음은 `seaborn`이 제공하는 주요 분포도 함수이다.

- `histplot()`: 히스토그램
- `kdeplot()`: 커널밀도추정 함수 그래프
- `displot()`: 분포도
- `rugplot()`: 러그플롯


### 히스토그램(histoplot)

- 수치형 데이터의 구간별 빈도수를 나타내는 그래프






```python
sns.histplot(data = tips, x = 'tip')
```

```python
sns.histplot(data = tips, x = 'tip', bins = 10)
```

```python
#빈도를 특정 범주별로 구분해서 plot
sns.histplot(data = tips, x = 'tip', hue = 'sex') 
```

```python
# 포개지 않고 누적
sns.histplot(data = tips, x = 'tip', hue = 'sex', multiple = 'stack') 
```

### 커널밀도추정 함수 그래프(kdeplot)

- 히스토그램을 매끄럽게 곡선으로 연결한 그래프

```python
sns.kdeplot(data = tips, x = 'tip')
```

```python
sns.kdeplot(data = tips, x = 'tip', hue = 'sex', multiple = 'stack')
```

### 분포도(displot)

- 분포도는 수치형 데이터 하나의 분포를 나타내는 그래프
- 캐글에서 분포도를 그릴 때는 `displot()`을 많이 사용함

```python
sns.displot(data = tips, x = 'tip')
```

```python
sns.displot(data = tips, x = 'tip', bins = 10, kde = True)
```

### 러그플롯(rugplot)

- 주변 분포(marginal distribution)을 나타내는 그래프
- 주로 다른 분포도 그래프와 함께 사용


```python
sns.kdeplot(data = tips, x = 'tip')
sns.rugplot(data = tips, x = 'tip')
```

## 4-4. 범주형 데이터 시각화

- 막대그래프
- 포인트플롯
- 박스플롯
- 바이올린플롯
- 카운트플롯


### 막대그래프(barplot)

- `barplot()`는 범주형 데이터에 따른 수치형 데이터의 평균과 신뢰구간을 구해준다
    - 수치형 데이터 평균은 막대 높이로, 신뢰구간은 오차 막대로 표현
    - x: 범주 / y: 수치
    

```python
sns.barplot(x = 'size', y = 'tip', data =tips)
```

### 포인트플롯(pointplot)

- 막대 그래프와 모양만 다를 뿐 동일한 정보를 제공

```python
sns.pointplot(x = 'size', y = 'tip', data =tips)
```

### 박스플롯(boxplot)

```python
sns.boxplot(x = 'size', y = 'tip', data =tips)
```

### 바이올린플롯(violinplot)


```python
sns.violinplot(x = 'size', y = 'tip', data =tips)
```

```python
sns.violinplot(x = 'size', y = 'tip', hue = 'sex', data =tips, split = True)
```

### 카운트플롯(countplot)

```python
sns.countplot(x = 'size', data =tips)
```

```python
sns.countplot(y = 'size', data =tips)
```

### 파이 그래프(pie)

- `seaborn`에서는 파이 그래프를 지원하지 않아 `matplotlib`을 사용해야 합니다

```python
import matplotlib
import matplotlib.pyplot as plt

x = [10, 60, 30]
labels = ['A', 'B', 'C']

plt.pie(x = x, labels = labels, autopct = '%.1f%%')
```

## 4-5. 데이터 관계 시각화

- 관계도는 여러 사이의 관계를 살펴보기 위한 그래프
    - 히트맵
    - 라인플롯
    - 산점도
    - 회귀선을 포함한 산점도
    
    
### 히트맵(heatmap)

- 데이터 간 관계를 색상으로 표현한 그래프
- 비교해야 할 데이터가 많을 때 주로 사용하며, `heatmap()` 함수를 이용

```python
import seaborn as sns

flights = sns.load_dataset('flights')
```

```python
flights.head()
```

`pivot()`함수는 index와 columns 파라미터에 전달한 피처를 각각 행과 열로 지정하고, values 파라미터에 전달한 피처를 합한 표를 반환

```python
flights_pivot = flights.pivot(index = 'month',
                             columns = 'year',
                             values = 'passengers')
```

```python
flights_pivot
```

```python
# 숫자로만 보니까 추이를 한눈에 파악하기 힘들다.
# 이럴 때 사용하는 그래프가 히트맵
sns.heatmap(data = flights_pivot)
```

### 라인플롯(lineplot)
- 두 수치형 데이터 사이의 관계를 나타낼 때 사용
- 기본적으로 x 파라미터에 전달한 값에 따라 y 파라미터에 전달한 값의 평균과 95% 신뢰구간을 나타냄

```python
sns.lineplot(x= 'year', y='passengers', data = flights)
# 해가 갈수록 평균 승객수도 많아짐
```

### 산점도(scatterplot)

- 산점도는 두 데이터의 관계를 점으로 표현하는 그래프

```python
sns.scatterplot(x='total_bill', y='tip', data = tips) #대체로 총액이 늘면 팁도 따라서 는다
```

```python
sns.scatterplot(x='total_bill', y='tip', hue = 'time', data = tips) # 점심과 저녁의 전체적인 추이는 비슷해보인다.
```

### 회귀선를 포함한 산점도 그래프(regplot)


```python
sns.regplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.regplot(x = 'total_bill', y = 'tip', ci = 99, data = tips) #신뢰구간 99%로 늘리면
```
