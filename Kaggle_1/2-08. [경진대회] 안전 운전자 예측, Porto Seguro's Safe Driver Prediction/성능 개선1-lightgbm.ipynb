# 4. ì•ˆì „ ìš´ì „ì ì˜ˆì¸¡ ê²½ì§„ëŒ€íšŒ ì„±ëŠ¥ ê°œì„  I : LightGBM ëª¨ë¸

- ë² ì´ìŠ¤ë¼ì¸ ëª¨ë¸ê³¼ ê°™ì€ LightGBMì„ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•˜ë©´ì„œ í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§ê³¼ í•˜ì´í¼íŒŒë¼ë¯¸í„° ìµœì í™” ì¶”ê°€ë¡œ ì ìš©

```python
import pandas as pd

# ë°ì´í„° ê²½ë¡œ
data_path = '/kaggle/input/porto-seguro-safe-driver-prediction/'

train = pd.read_csv(data_path + 'train.csv', index_col='id')
test = pd.read_csv(data_path + 'test.csv', index_col='id')
submission = pd.read_csv(data_path + 'sample_submission.csv', index_col='id')
```

## 4-1. í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§

### 4-1-1. ë°ì´í„° í•©ì¹˜ê¸°

```python
all_data = pd.concat([train, test], ignore_index=True)
all_data = all_data.drop('target', axis=1) # íƒ€ê¹ƒê°’ ì œê±°

all_features = all_data.columns # ì „ì²´ í”¼ì²˜
```

### 4-1-2. ëª…ëª©ì  í”¼ì²˜ ì›-í•« ì¸ì½”ë”©

```python
from sklearn.preprocessing import OneHotEncoder

# ëª…ëª©í˜• í”¼ì²˜
cat_features = [feature for feature in all_features if 'cat' in feature] 

# ì›-í•« ì¸ì½”ë”© ì ìš©
onehot_encoder = OneHotEncoder()
encoded_cat_matrix = onehot_encoder.fit_transform(all_data[cat_features]) 
```

### 4-1-3. íŒŒìƒ í”¼ì²˜ ì¶”ê°€

- EDAì—ì„œëŠ” í•„ìš” ì—†ëŠ” í”¼ì²˜ë¥¼ ì¶”ë¦¬ëŠ” ê²ƒ ì™¸ì— íŠ¹ë³„í•œ í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§ ìš”ì†Œë¥¼ ì°¾ì•„ë‚´ì§€ ëª»í•¨.
- ì„±ëŠ¥ì„ ë” ë†’ì´ê¸° ìœ„í•´ ì‹œë„ í•  ìˆ˜ ìˆëŠ” ëª‡ ê°€ì§€ ë°©ë²• ì‚´í”¼ì


### 1) **í•œ ë°ì´í„°ê°€ ê°€ì§„ ê²°ì¸¡ê°’ ê°œìˆ˜ë¥¼ íŒŒìƒ í”¼ì²˜ë¡œ ë§Œë“¤ì**

```python
# 'ë°ì´í„° í•˜ë‚˜ë‹¹ ê²°ì¸¡ê°’ ê°œìˆ˜'ë¥¼ íŒŒìƒ í”¼ì²˜ë¡œ ì¶”ê°€
all_data['num_missing'] = (all_data==-1).sum(axis=1)
```

```python
# ëª…ëª©í˜• í”¼ì²˜, calc ë¶„ë¥˜ì˜ í”¼ì²˜ë¥¼ ì œì™¸í•œ í”¼ì²˜
remaining_features = [feature for feature in all_features
                      if ('cat' not in feature and 'calc' not in feature)] 
# num_missingì„ remaining_featuresì— ì¶”ê°€
remaining_features.append('num_missing')
```

- ì´ë¡œì„œ `all_data`ì™€ `remaining_features`ì— ìƒˆë¡œìš´ íŒŒìƒ í”¼ì²˜ `num_missing`ì´ ì¶”ê°€ë¨

### **2) ëª¨ë“  `ind` í”¼ì²˜ ê°’ì„ ì—°ê²°í•´ì„œ ìƒˆë¡œìš´ í”¼ì²˜ë¥¼ ë§Œë“¤ì**

- `mix_ind`
- ë’¤ì´ì–´ ì¶”ê°€í•  'ëª…ëª©í˜• í”¼ì²˜ì˜ ê³ ìœ³ê°’ë³„ ê°œìˆ˜' í”¼ì²˜ë¥¼ ë§Œë“¤ ë•Œ ì‚¬ìš©í•˜ê¸° ìœ„í•œ ì„ì‹œ í”¼ì²˜

```python
# ë¶„ë¥˜ê°€ indì¸ í”¼ì²˜
ind_features = [feature for feature in all_features if 'ind' in feature]

is_first_feature = True
for ind_feature in ind_features:
    if is_first_feature:
        all_data['mix_ind'] = all_data[ind_feature].astype(str) + '_'
        is_first_feature = False
    else:
        all_data['mix_ind'] += all_data[ind_feature].astype(str) + '_'
```

```python
all_data['mix_ind']
```

### **3) ëª…ëª©í˜• í”¼ì²˜ì˜ ê³ ìœ³ê°’ë³„ ê°œìˆ˜ë¥¼ ìƒˆë¡œìš´ í”¼ì²˜ë¡œ ì¶”ê°€**


```python
cat_count_features = []
for feature in cat_features+['mix_ind']:
    val_counts_dict = all_data[feature].value_counts().to_dict()
    all_data[f'{feature}_count'] = all_data[feature].apply(lambda x: 
                                                           val_counts_dict[x])
    cat_count_features.append(f'{feature}_count')
```

```python
cat_count_features
```

- ì§€ê¸ˆê¹Œì§€ ë§Œë“  í”¼ì²˜ëŠ” ë‹¤ìŒê³¼ ê°™ë‹¤
    - `encoded_cat_matrix`: ì›-í•« ì¸ì½”ë”©ëœ ëª…ëª©í˜• í”¼ì²˜
    - `remaining_features`: ëª…ëª©í˜• í”¼ì²˜ì™€ calc ë¶„ë¥˜ì˜ í”¼ì²˜ë¥¼ ì œì™¸í•œ í”¼ì²˜ë“¤ (+ `num_missing`)
    - `cat_count_features`: mix_ind ë¥¼ í¬í•¨í•œ ëª…ëª©í˜• í”¼ëŸ¬ì˜ ê³ ìœ³ê°’ë³„ ê°œìˆ˜ íŒŒìƒ í”¼ì²˜




### 4-1-4. í•„ìš” ì—†ëŠ” í”¼ì²˜ ì œê±°

- íƒìƒ‰ì  ë°ì´í„° ë¶„ì„ì—ì„œ í•„ìš” ì—†ë‹¤ê³  íŒë‹¨í•œ í”¼ì²˜ëŠ” ì œê±°í•œ ë‹¤ìŒ
- ì§€ê¸ˆê¹Œì§€ í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§í•œ ëª¨ë“  ë°ì´í„°ë¥¼ í•©ì¹  ê²ƒ

```python
from scipy import sparse
# í•„ìš” ì—†ëŠ” í”¼ì²˜ë“¤
drop_features = ['ps_ind_14', 'ps_ind_10_bin', 'ps_ind_11_bin', 
                 'ps_ind_12_bin', 'ps_ind_13_bin', 'ps_car_14']

# remaining_features, cat_count_featuresì—ì„œ drop_featuresë¥¼ ì œê±°í•œ ë°ì´í„°
all_data_remaining = all_data[remaining_features+cat_count_features].drop(drop_features, axis=1)

# ë°ì´í„° í•©ì¹˜ê¸°
all_data_sprs = sparse.hstack([sparse.csr_matrix(all_data_remaining),
                               encoded_cat_matrix],
                              format='csr')
```

### ğŸ“– ì§€ê¸ˆê¹Œì§€ì˜ í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§ìœ¼ë¡œ ìˆ˜í–‰í•œ ì¼ë“¤ì„ ì •ë¦¬í•˜ë©´ ë‹¤ìŒê³¼ ê°™ë‹¤ 

- ëª…ëª©í˜• í”¼ì²˜ì— ì›-í•« ì¸ì½”ë”©ì„ ì ìš©
- ë°ì´í„° í•˜ë‚˜ë‹¹ ê°€ì§€ê³  ìˆëŠ” ê²°ì¸¡ê°’ ê°œìˆ˜ë¥¼ ìƒˆë¡œìš´ í”¼ì²˜ë¡œ ë§Œë“¤ì—ˆë‹¤.
- ëª¨ë“  ind í”¼ì²˜ ê°’ì„ ì—°ê²°í•´ì„œ ìƒˆë¡œìš´ ëª…ëª©í˜• í”¼ì²˜ë¥¼ ë§Œë“¤ì—ˆë‹¤.
- ëª…ëª©í˜• í”¼ì²˜ì˜ ê³ ìœ³ê°’ë³„ ê°œìˆ˜ë¥¼ ìƒˆë¡œìš´ í”¼ì²˜ë¡œ ë§Œë“¤ì—ˆë‹¤.
- í•„ìš” ì—†ëŠ” í”¼ì²˜ë¥¼ ì œê±° (drop_featuresì™€ calc ë¶„ë¥˜ì˜ í”¼ì²˜ë“¤)

### 4-1-5. ë°ì´í„° ë‚˜ëˆ„ê¸°

```python
num_train = len(train) # í›ˆë ¨ ë°ì´í„° ê°œìˆ˜

# í›ˆë ¨ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„° ë‚˜ëˆ„ê¸°
X = all_data_sprs[:num_train]
X_test = all_data_sprs[num_train:]

y = train['target'].values
```

```python
import numpy as np

def eval_gini(y_true, y_pred):
    # ì‹¤ì œê°’ê³¼ ì˜ˆì¸¡ê°’ì˜ í¬ê¸°ê°€ ê°™ì€ì§€ í™•ì¸ (ê°’ì´ ë‹¤ë¥´ë©´ ì˜¤ë¥˜ ë°œìƒ)
    assert y_true.shape == y_pred.shape

    n_samples = y_true.shape[0]                      # ë°ì´í„° ê°œìˆ˜
    L_mid = np.linspace(1 / n_samples, 1, n_samples) # ëŒ€ê°ì„  ê°’

    # 1) ì˜ˆì¸¡ê°’ì— ëŒ€í•œ ì§€ë‹ˆê³„ìˆ˜
    pred_order = y_true[y_pred.argsort()] # y_pred í¬ê¸°ìˆœìœ¼ë¡œ y_true ê°’ ì •ë ¬
    L_pred = np.cumsum(pred_order) / np.sum(pred_order) # ë¡œë Œì¸  ê³¡ì„ 
    G_pred = np.sum(L_mid - L_pred)       # ì˜ˆì¸¡ ê°’ì— ëŒ€í•œ ì§€ë‹ˆê³„ìˆ˜

    # 2) ì˜ˆì¸¡ì´ ì™„ë²½í•  ë•Œ ì§€ë‹ˆê³„ìˆ˜
    true_order = y_true[y_true.argsort()] # y_true í¬ê¸°ìˆœìœ¼ë¡œ y_true ê°’ ì •ë ¬
    L_true = np.cumsum(true_order) / np.sum(true_order) # ë¡œë Œì¸  ê³¡ì„ 
    G_true = np.sum(L_mid - L_true)       # ì˜ˆì¸¡ì´ ì™„ë²½í•  ë•Œ ì§€ë‹ˆê³„ìˆ˜

    # ì •ê·œí™”ëœ ì§€ë‹ˆê³„ìˆ˜
    return G_pred / G_true

# LightGBMìš© gini() í•¨ìˆ˜
def gini(preds, dtrain):
    labels = dtrain.get_label()
    return 'gini', eval_gini(labels, preds), True # ë°˜í™˜ê°’
```

## 4-2. í•˜ì´í¼íŒŒë¼ë¯¸í„° ìµœì í™”

- ì„±ëŠ¥ì´ ìš°ìˆ˜í•œ ëª¨ë¸ì„ ë§Œë“¤ê¸° ìœ„í•´ì„œ í•„ìš”
- **ë² ì´ì§€ì•ˆ ìµœì í™” ê¸°ë²•**ì„ í™œìš©í•´ í•˜ì´í¼íŒŒë¼ë¯¸í„°ë¥¼ ì¡°ì •



### 4-2-1. ë°ì´í„°ì…‹ ì¤€ë¹„

- ë² ì´ì§€ì•ˆ ìµœì í™”ë¥¼ ìœ„í•œ ë°ì´í„°ì…‹ì„ ë§Œë“¤ì

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split

# 8:2 ë¹„ìœ¨ë¡œ í›ˆë ¨ ë°ì´í„°, ê²€ì¦ ë°ì´í„° ë¶„ë¦¬ (ë² ì´ì§€ì•ˆ ìµœì í™” ìˆ˜í–‰ìš©)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, 
                                                      test_size=0.2, 
                                                      random_state=0)

# ë² ì´ì§€ì•ˆ ìµœì í™”ìš© ë°ì´í„°ì…‹
bayes_dtrain = lgb.Dataset(X_train, y_train)
bayes_dvalid = lgb.Dataset(X_valid, y_valid)
```

- ì´ë ‡ê²Œ ë‚˜ë‰œ í›ˆë ¨ ë°ì´í„°ì™€ ê²€ì¦ ë°ì´í„°ë¥¼ í™œìš©í•´ ë² ì´ì§€ì•ˆ ìµœì í™”ìš© ë°ì´í„°ì…‹ì„ ë§Œë“¤ì


### 4-2-2. í•˜ì´í¼íŒŒë¼ë¯¸í„° ë²”ìœ„ ì„¤ì •

1. í•˜ì´í¼íŒŒë¼ë¯¸í„° ë²”ìœ„ë¥¼ ì ì  ì¢íˆëŠ” ë°©ë²•
2. ë‹¤ë¥¸ ìƒìœ„ê¶Œ ìºê¸€ëŸ¬ê°€ ì„¤ì •í•œ í•˜ì´í¼íŒŒë¼ë¯¸í„°ë¥¼ ì°¸ê³ í•˜ëŠ” ë°©ë²•

```python
# ë² ì´ì§€ì•ˆ ìµœì í™”ë¥¼ ìœ„í•œ í•˜ì´í¼íŒŒë¼ë¯¸í„° ë²”ìœ„
param_bounds = {'num_leaves': (30, 40),
                'lambda_l1': (0.7, 0.9),
                'lambda_l2': (0.9, 1),
                'feature_fraction': (0.6, 0.7),
                'bagging_fraction': (0.6, 0.9),
                'min_child_samples': (6, 10),
                'min_child_weight': (10, 40)}

# ê°’ì´ ê³ ì •ëœ í•˜ì´í¼íŒŒë¼ë¯¸í„°
fixed_params = {'objective': 'binary',
                'learning_rate': 0.005,
                'bagging_freq': 1,
                'force_row_wise': True,
                'random_state': 1991}
```

### 4-2-3. (ë² ì´ì§€ì•ˆ ìµœì í™”ìš©) í‰ê°€ì§€í‘œ ê³„ì‚° í•¨ìˆ˜ ì‘ì„±

```python
def eval_function(num_leaves, lambda_l1, lambda_l2, feature_fraction,
                  bagging_fraction, min_child_samples, min_child_weight):
    '''ìµœì í™”í•˜ë ¤ëŠ” í‰ê°€ì§€í‘œ(ì§€ë‹ˆê³„ìˆ˜) ê³„ì‚° í•¨ìˆ˜'''
    
    # ë² ì´ì§€ì•ˆ ìµœì í™”ë¥¼ ìˆ˜í–‰í•  í•˜ì´í¼íŒŒë¼ë¯¸í„° 
    params = {'num_leaves': int(round(num_leaves)),
              'lambda_l1': lambda_l1,
              'lambda_l2': lambda_l2,
              'feature_fraction': feature_fraction,
              'bagging_fraction': bagging_fraction,
              'min_child_samples': int(round(min_child_samples)),
              'min_child_weight': min_child_weight,
              'feature_pre_filter': False}
    # ê³ ì •ëœ í•˜ì´í¼íŒŒë¼ë¯¸í„°ë„ ì¶”ê°€
    params.update(fixed_params)
    
    print('í•˜ì´í¼íŒŒë¼ë¯¸í„°:', params)    
    
    # LightGBM ëª¨ë¸ í›ˆë ¨
    lgb_model = lgb.train(params=params, 
                           train_set=bayes_dtrain,
                           num_boost_round=2500,
                           valid_sets=bayes_dvalid,
                           feval=gini,
                           early_stopping_rounds=300,
                           verbose_eval=False)
    # ê²€ì¦ ë°ì´í„°ë¡œ ì˜ˆì¸¡ ìˆ˜í–‰
    preds = lgb_model.predict(X_valid) 
    # ì§€ë‹ˆê³„ìˆ˜ ê³„ì‚°
    gini_score = eval_gini(y_valid, preds)
    print(f'ì§€ë‹ˆê³„ìˆ˜ : {gini_score}\n')
    
    return gini_score
```

### 4-2-4. ìµœì í™” ìˆ˜í–‰

```python
from bayes_opt import BayesianOptimization

# ë² ì´ì§€ì•ˆ ìµœì í™” ê°ì²´ ìƒì„±
optimizer = BayesianOptimization(f=eval_function,      # í‰ê°€ì§€í‘œ ê³„ì‚° í•¨ìˆ˜
                                 pbounds=param_bounds, # í•˜ì´í¼íŒŒë¼ë¯¸í„° ë²”ìœ„
                                 random_state=0)
```

```python
# ë² ì´ì§€ì•ˆ ìµœì í™” ìˆ˜í–‰
optimizer.maximize(init_points=3, n_iter=6)
```

### 4-2-5. ê²°ê³¼ í™•ì¸

```python
# í‰ê°€í•¨ìˆ˜ ì ìˆ˜ê°€ ìµœëŒ€ì¼ ë•Œ í•˜ì´í¼íŒŒë¼ë¯¸í„°
max_params = optimizer.max['params']
max_params
```

```python
# ì •ìˆ˜í˜• í•˜ì´í¼íŒŒë¼ë¯¸í„° ë³€í™˜
max_params['num_leaves'] = int(round(max_params['num_leaves']))
max_params['min_child_samples'] = int(round(max_params['min_child_samples']))
```

```python
# ê°’ì´ ê³ ì •ëœ í•˜ì´í¼íŒŒë¼ë¯¸í„° ì¶”ê°€
max_params.update(fixed_params)
```

## 4-3. ëª¨ë¸ í›ˆë ¨ ë° ì„±ëŠ¥ ê²€ì¦

- ë² ì´ì§€ì•ˆ ìµœì í™”ë¡œ ì°¾ì€ í•˜ì´í¼íŒŒë¼ë¯¸í„°ë¥¼ í™œìš©í•´ `LightGBM` ëª¨ë¸ì„ ë‹¤ì‹œ í›ˆë ¨í•´ë³´ì

```python
from sklearn.model_selection import StratifiedKFold

# ì¸µí™” K í´ë“œ êµì°¨ ê²€ì¦ê¸° ìƒì„±
folds = StratifiedKFold(n_splits=5, shuffle=True, random_state=1991)

# OOF ë°©ì‹ìœ¼ë¡œ í›ˆë ¨ëœ ëª¨ë¸ë¡œ ê²€ì¦ ë°ì´í„° íƒ€ê¹ƒê°’ì„ ì˜ˆì¸¡í•œ í™•ë¥ ì„ ë‹´ì„ 1ì°¨ì› ë°°ì—´
oof_val_preds = np.zeros(X.shape[0]) 
# OOF ë°©ì‹ìœ¼ë¡œ í›ˆë ¨ëœ ëª¨ë¸ë¡œ í…ŒìŠ¤íŠ¸ ë°ì´í„° íƒ€ê¹ƒê°’ì„ ì˜ˆì¸¡í•œ í™•ë¥ ì„ ë‹´ì„ 1ì°¨ì› ë°°ì—´
oof_test_preds = np.zeros(X_test.shape[0]) 

# OOF ë°©ì‹ìœ¼ë¡œ ëª¨ë¸ í›ˆë ¨, ê²€ì¦, ì˜ˆì¸¡
for idx, (train_idx, valid_idx) in enumerate(folds.split(X, y)):
    # ê° í´ë“œë¥¼ êµ¬ë¶„í•˜ëŠ” ë¬¸êµ¬ ì¶œë ¥
    print('#'*40, f'í´ë“œ {idx+1} / í´ë“œ {folds.n_splits}', '#'*40)
    
    # í›ˆë ¨ìš© ë°ì´í„°, ê²€ì¦ìš© ë°ì´í„° ì„¤ì •
    X_train, y_train = X[train_idx], y[train_idx] # í›ˆë ¨ìš© ë°ì´í„°
    X_valid, y_valid = X[valid_idx], y[valid_idx] # ê²€ì¦ìš© ë°ì´í„°

    # LightGBM ì „ìš© ë°ì´í„°ì…‹ ìƒì„±
    dtrain = lgb.Dataset(X_train, y_train) # LightGBM ì „ìš© í›ˆë ¨ ë°ì´í„°ì…‹
    dvalid = lgb.Dataset(X_valid, y_valid) # LightGBM ì „ìš© ê²€ì¦ ë°ì´í„°ì…‹
                          
    # LightGBM ëª¨ë¸ í›ˆë ¨
    lgb_model = lgb.train(params=max_params,    # ìµœì  í•˜ì´í¼íŒŒë¼ë¯¸í„°
                          train_set=dtrain,     # í›ˆë ¨ ë°ì´í„°ì…‹
                          num_boost_round=2500, # ë¶€ìŠ¤íŒ… ë°˜ë³µ íšŸìˆ˜
                          valid_sets=dvalid,    # ì„±ëŠ¥ í‰ê°€ìš© ê²€ì¦ ë°ì´í„°ì…‹
                          feval=gini,           # ê²€ì¦ìš© í‰ê°€ì§€í‘œ
                          early_stopping_rounds=300, # ì¡°ê¸°ì¢…ë£Œ ì¡°ê±´
                          verbose_eval=100)     # 100ë²ˆì§¸ë§ˆë‹¤ ì ìˆ˜ ì¶œë ¥
    
    # í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¥¼ í™œìš©í•´ OOF ì˜ˆì¸¡
    oof_test_preds += lgb_model.predict(X_test)/folds.n_splits
    # ëª¨ë¸ ì„±ëŠ¥ í‰ê°€ë¥¼ ìœ„í•œ ê²€ì¦ ë°ì´í„° íƒ€ê¹ƒê°’ ì˜ˆì¸¡ 
    oof_val_preds[valid_idx] += lgb_model.predict(X_valid)
    
    # ê²€ì¦ ë°ì´í„° ì˜ˆì¸¡í™•ë¥ ì— ëŒ€í•œ ì •ê·œí™” ì§€ë‹ˆê³„ìˆ˜
    gini_score = eval_gini(y_valid, oof_val_preds[valid_idx])
    print(f'í´ë“œ {idx+1} ì§€ë‹ˆê³„ìˆ˜ : {gini_score}\n')
```

```python
print('OOF ê²€ì¦ ë°ì´í„° ì§€ë‹ˆê³„ìˆ˜ :', eval_gini(y, oof_val_preds))
```

## 4-4. ì˜ˆì¸¡ ë° ê²°ê³¼ ì œì¶œ

```python
submission['target'] = oof_test_preds
submission.to_csv('submission.csv')
```
