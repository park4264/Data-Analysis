## 1. í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§ I : í”¼ì²˜ëª… í•œê¸€í™”ì™€ ë°ì´í„° ë‹¤ìš´ìºìŠ¤íŒ…

```python
sales_train = sales_train.rename(columns={'date': 'ë‚ ì§œ', 
                                          'date_block_num': 'ì›”ID',
                                          'shop_id': 'ìƒì ID',
                                          'item_id': 'ìƒí’ˆID',
                                          'item_price': 'íŒë§¤ê°€',
                                          'item_cnt_day': 'íŒë§¤ëŸ‰'})

shops = shops.rename(columns={'shop_name': 'ìƒì ëª…',
                              'shop_id': 'ìƒì ID'})

items = items.rename(columns={'item_name': 'ìƒí’ˆëª…',
                              'item_id': 'ìƒí’ˆID',
                              'item_category_id': 'ìƒí’ˆë¶„ë¥˜ID'})

item_categories = item_categories.rename(columns=
                                         {'item_category_name': 'ìƒí’ˆë¶„ë¥˜ëª…',
                                          'item_category_id': 'ìƒí’ˆë¶„ë¥˜ID'})

test = test.rename(columns={'shop_id': 'ìƒì ID',
                            'item_id': 'ìƒí’ˆID'})
```

```python
def downcast(df, verbose=True):
    start_mem = df.memory_usage().sum() / 1024**2
    for col in df.columns:
        dtype_name = df[col].dtype.name
        if dtype_name == 'object':
            pass
        elif dtype_name == 'bool':
            df[col] = df[col].astype('int8')
        elif dtype_name.startswith('int') or (df[col].round() == df[col]).all():
            df[col] = pd.to_numeric(df[col], downcast='integer')
        else:
            df[col] = pd.to_numeric(df[col], downcast='float')
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose:
        print('{:.1f}% ì••ì¶•ë¨'.format(100 * (start_mem - end_mem) / start_mem))
    
    return df

all_df = [sales_train, shops, items, item_categories, test]
for df in all_df:
    df = downcast(df)
```

## 3. í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§ III : ë°ì´í„° ì¡°í•© ë° íŒŒìƒ í”¼ì²˜ ìƒì„±

- ê·¸ ë‹¤ìŒ ì›”ê°„ íŒë§¤ëŸ‰, í‰ê·  íŒë§¤ê°€, íŒë§¤ê±´ìˆ˜ í”¼ì²˜ë¥¼ ë§Œë“ ë‹¤.


### 3-1. ë°ì´í„° ì¡°í•©

```python
from itertools import product

train = []
# ì›”ID, ìƒì ID, ìƒí’ˆID ì¡°í•© ìƒì„±
for i in sales_train['ì›”ID'].unique():
    all_shop = sales_train.loc[sales_train['ì›”ID']==i, 'ìƒì ID'].unique()
    all_item = sales_train.loc[sales_train['ì›”ID']==i, 'ìƒí’ˆID'].unique()
    train.append(np.array(list(product([i], all_shop, all_item))))

idx_features = ['ì›”ID', 'ìƒì ID', 'ìƒí’ˆID'] # ê¸°ì¤€ í”¼ì²˜
train = pd.DataFrame(np.vstack(train), columns=idx_features)
```

### 3-2. íŒŒìƒ í”¼ì²˜ ìƒì„±

- ì´ë²ˆì—ëŠ” ì´ 3ê°œì˜ í”¼ì²˜ë¥¼ ìƒˆë¡œ ë§Œë“¤ê² ë‹¤.
- ì›”ID, ìƒì ID, ìƒí’ˆID ë³„ 'ì›”ê°„ íŒë§¤ëŸ‰', 'í‰ê·  íŒë§¤ê°€' í”¼ì²˜ë¥¼ ë§Œë“¤ì

```python
group = sales_train.groupby(idx_features).agg({'íŒë§¤ëŸ‰': 'sum',
                                               'íŒë§¤ê°€': 'mean'})
group = group.reset_index()
group = group.rename(columns={'íŒë§¤ëŸ‰': 'ì›”ê°„ íŒë§¤ëŸ‰', 'íŒë§¤ê°€': 'í‰ê·  íŒë§¤ê°€'})

train = train.merge(group, on=idx_features, how='left')

train.head()
```

- ê²°ì¸¡ê°’ì€ 0ìœ¼ë¡œ ëŒ€ì²´

```python
import gc

# group ë³€ìˆ˜ ê°€ë¹„ì§€ ì»¬ë ‰ì…˜
del group
gc.collect();
```

- ê·¸ ë‹¤ìŒìœ¼ë¡œëŠ” 'ê¸°ì¤€ í”¼ì²˜ë³„ ìƒí’ˆ íŒë§¤ê±´ìˆ˜' íŒŒìƒ í”¼ì²˜ë¥¼ ë§Œë“¤ì

```python
# ìƒí’ˆ íŒë§¤ê±´ìˆ˜ í”¼ì²˜ ì¶”ê°€
group = sales_train.groupby(idx_features).agg({'íŒë§¤ëŸ‰': 'count'})
group = group.reset_index()
group = group.rename(columns={'íŒë§¤ëŸ‰': 'íŒë§¤ê±´ìˆ˜'})

train = train.merge(group, on=idx_features, how='left')

# ê°€ë¹„ì§€ ì»¬ë ‰ì…˜
del group, sales_train
gc.collect()

train.head()
```

### 4-1. í…ŒìŠ¤íŠ¸ ë°ì´í„° ì´ì–´ë¶™ì´ê¸°

### 4-2. ëª¨ë“  ë°ì´í„° ë³‘í•©

```python
# ë‚˜ë¨¸ì§€ ë°ì´í„° ë³‘í•©
all_data = all_data.merge(shops, on='ìƒì ID', how='left')
all_data = all_data.merge(items, on='ìƒí’ˆID', how='left')
all_data = all_data.merge(item_categories, on='ìƒí’ˆë¶„ë¥˜ID', how='left')

# ë°ì´í„° ë‹¤ìš´ìºìŠ¤íŒ…
all_data = downcast(all_data)
```

- ê¸°ì¤€ í”¼ì²˜ë¡œ ê·¸ë£¹í™”í•´ ì›”ê°„ í‰ê·  íŒë§¤ëŸ‰ì„ êµ¬í•´ì£¼ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“¤ì–´ë³´ì
- íŒŒë¼ë¯¸í„°ëŠ” ì´ 3ê°œ
    - `df`: ì‘ì—…í•  ì „ì²´ ë°ì´í„° (DataFrame)
    - `mean_features`: ìƒˆë¡œ ë§Œë“  ì›”ê°„ í‰ê·  íŒë§¤ëŸ‰ íŒŒìƒ í”¼ì²˜ëª…ì„ ì €ì¥í•˜ëŠ” ë¦¬ìŠ¤íŠ¸
    - `idx_features`: ê¸°ì¤€ í”¼ì²˜

```python
def add_mean_features(df, mean_features, idx_features):
    # ê¸°ì¤€ í”¼ì²˜ í™•ì¸ 
    assert (idx_features[0] == 'ì›”ID') and \
           len(idx_features) in [2, 3]
    
    # íŒŒìƒ í”¼ì²˜ëª… ì„¤ì • 
    if len(idx_features) == 2:
        feature_name = idx_features[1] + 'ë³„ í‰ê·  íŒë§¤ëŸ‰'
    else:
        feature_name = idx_features[1] + ' ' + idx_features[2] + 'ë³„ í‰ê·  íŒë§¤ëŸ‰'
    
    # ê¸°ì¤€ í”¼ì²˜ë¥¼ í† ëŒ€ë¡œ ê·¸ë£¹í™”í•´ ì›”ê°„ í‰ê·  íŒë§¤ëŸ‰ êµ¬í•˜ê¸° 
    group = df.groupby(idx_features).agg({'ì›”ê°„ íŒë§¤ëŸ‰': 'mean'})
    group = group.reset_index()
    group = group.rename(columns={'ì›”ê°„ íŒë§¤ëŸ‰': feature_name})
    
    # dfì™€ group ë³‘í•© 
    df = df.merge(group, on=idx_features, how='left')
    # ë°ì´í„° ë‹¤ìš´ìºìŠ¤íŒ… 
    df = downcast(df, verbose=False)
    # ìƒˆë¡œ ë§Œë“  feature_name í”¼ì²˜ëª…ì„ mean_features ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€ 
    mean_features.append(feature_name)
    
    # ê°€ë¹„ì§€ ì»¬ë ‰ì…˜
    del group
    gc.collect()
    
    return df, mean_features
```

- ì–´ë–¤ í”¼ì²˜ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ê·¸ë£¹í™”í•´ì•¼ ì˜ˆì¸¡ ì„±ëŠ¥ì´ ì¢‹ì•„ì§€ëŠ”ì§€ í™•ì‹¤íˆ ì•Œ ìˆ˜ëŠ” ì—†ë‹¤

### 5-1. ì‹œì°¨ í”¼ì²˜ ìƒì„± ì›ë¦¬ ë° í•¨ìˆ˜ êµ¬í˜„

- ì§€ê¸ˆê¹Œì§€ ë§Œë“  ê¸°ì¤€ í”¼ì²˜ë³„ ì›”ê°„ í‰ê·  íŒë§¤ëŸ‰ í”¼ì²˜ëŠ” ë‹¤ìŒê³¼ ê°™ë‹¤.
    1. {ìƒí’ˆID}ë³„ í‰ê·  íŒë§¤ëŸ‰
    2. {ìƒí’ˆID + ë„ì‹œ}ë³„ í‰ê·  íŒë§¤ëŸ‰
    3. {ìƒí’ˆID + ìƒí’ˆë²”ì£¼ID}ë³„ í‰ê·  íŒë§¤ëŸ‰
- ì´ì œ ì´ ì„¸ ì§•ê²€ë‹¤ë¦¬ í”¼ì²˜ë¥¼ í™œìš©í•´ ì‹œì°¨ í”¼ì²˜ë¥¼ êµ¬í•´ë³´ê² ë‹¤.

<br>

- âœ¨ <u>**ì‹œì°¨ í”¼ì²˜ëŠ” ì‹œê³„ì—´ ë¬¸ì œì—ì„œ ìì£¼ ì‚¬ìš©í•˜ëŠ” íŒŒìƒ í”¼ì²˜**</u> âœ¨
    - í˜„ì‹œì  ë°ì´í„°ì— ê³¼ê±° ì‹œì  ë°ì´í„°ë¥¼ ì¶”ê°€í•œë‹¤ëŠ” ê°œë…
    - ê³¼ê±° ì‹œì  ë°ì´í„°ëŠ” í–¥í›„ íŒë§¤ëŸ‰ ì˜ˆì¸¡ì— ìœ ìš©
    - ì‹œì°¨ í”¼ì²˜ëŠ” í•œ ë‹¬ ì „, ë‘ ë‹¬ ì „, ì„¸ ë‹¬ ì „ ë“± ì›í•˜ëŠ” ì‹œì ê¹Œì§€ ìƒì„± ê°€ëŠ¥
    
<br>

- `add_lag_features()`ëŠ” ë°©ê¸ˆ ì„¤ëª…í•œ ì›ë¦¬ë¥¼ ì ìš©í•´ ë§Œë“  ì‹œì°¨ í”¼ì²˜ë¥¼ ì¶”ê°€í•˜ëŠ” í•¨ìˆ˜
- íŒŒë¼ë¯¸í„°ëŠ” ë‹¤ìŒê³¼ ê°™ë‹¤
    - `df`: ì›ë³¸ ë°ì´í„°
    - `lag_feature_to_clip`: 'ê°’ì˜ ë²”ìœ„ë¥¼ 0 ~ 20 ì‚¬ì´ë¡œ ì œí•œí•  í”¼ì²˜'fmf ekadmf fltmxm
    - `idx_feature`: ê¸°ì¤€ í”¼ì²˜
    - `lag_feature`: ì‹œì°¨ë¥¼ ë§Œë“¤ í”¼ì²˜
    - `nlags`: ì‹œì°¨
        - 1: í•œ ë‹¬ ì „ ì‹œì°¨ í”¼ì²˜ë§Œ ìƒì„±
        - 2: í•œ ë‹¬ ì „, ë‘ ë‹¬ ì „ ì‹œì°¨ í”¼ì²˜ ìƒì„±
        - 3: í•œ ë‹¬ ì „, ë‘ ë‹¬ ì „, ì„¸ ë‹¬ ì „ ì‹œì°¨ í”¼ì²˜ë¥¼ ëª¨ë‘ ìƒì„±
    - `clip`: ìƒˆë¡œ ë§Œë“  ì‹œì°¨ í”¼ì²˜ë¥¼ `lag_features_to_clip` ë¦¬ìŠ¤íŠ¸ì— ì €ì¥í• ì§€ ì—¬ë¶€







```python
def add_lag_features(df, lag_features_to_clip, idx_features, 
                     lag_feature, nlags=3, clip=False):
    # ì‹œì°¨ í”¼ì²˜ ìƒì„±ì— í•„ìš”í•œ DataFrame ë¶€ë¶„ë§Œ ë³µì‚¬ 
    df_temp = df[idx_features + [lag_feature]].copy() 

    # ì‹œì°¨ í”¼ì²˜ ìƒì„± 
    for i in range(1, nlags+1):
        # ì‹œì°¨ í”¼ì²˜ëª… 
        lag_feature_name = lag_feature +'_ì‹œì°¨' + str(i)
        # df_temp ì—´ ì´ë¦„ ì„¤ì • 
        df_temp.columns = idx_features + [lag_feature_name]
        # df_tempì˜ date_block_num í”¼ì²˜ì— 1 ë”í•˜ê¸° 
        df_temp['ì›”ID'] += 1
        # idx_featureë¥¼ ê¸°ì¤€ìœ¼ë¡œ dfì™€ df_temp ë³‘í•©í•˜ê¸° 
        df = df.merge(df_temp.drop_duplicates(), 
                      on=idx_features, 
                      how='left')
        # ê²°ì¸¡ê°’ 0ìœ¼ë¡œ ëŒ€ì²´ 
        df[lag_feature_name] = df[lag_feature_name].fillna(0)
        # 0 ~ 20 ì‚¬ì´ë¡œ ì œí•œí•  ì‹œì°¨ í”¼ì²˜ëª…ì„ lag_features_to_clipì— ì¶”ê°€ 
        if clip: 
            lag_features_to_clip.append(lag_feature_name)
    
    # ë°ì´í„° ë‹¤ìš´ìºìŠ¤íŒ…
    df = downcast(df, False)
    # ê°€ë¹„ì§€ ì»¬ë ‰ì…˜
    del df_temp
    gc.collect()
    
    return df, lag_features_to_clip
```
